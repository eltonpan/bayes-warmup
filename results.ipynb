```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from visualize import *
import pickle
```

```python
traj = get_trajectory('morgan', 'random', '20', return_mols=True)
```

```python
feat_dict = {}
for feat_name in ['morgan', 'molformer']:
    split_dict = {}
    for split in ['random', 'stratified']:
        n_sample_dict = {}
        for n_sample in ['5', '10', '20', '50', '100', '200']:
            traj, mols = get_trajectory(feat_name, split, n_sample, return_mols=True)
            n_sample_dict[n_sample] = {'traj': traj, 'mols': mols}
        split_dict[split] = n_sample_dict
    feat_dict[feat_name] = split_dict  
```

```python
with open('all_results.pkl', "wb") as f:
    pickle.dump(feat_dict, f)
with open('all_results.pkl', "rb") as f:
    feat_dict = pickle.load(f)
```

```python
def plot_mean_trajectory(feat_type, split, n_sample):

    traj = get_trajectory(feat_type, split, n_sample)

    plt.figure(figsize=(10,10))
    yerr = traj.std(0)
    plt.fill_between(range(50), traj.mean(0)-yerr, traj.mean(0) + yerr, color = 'red', alpha = 0.5)
    plt.plot(range(50), traj.mean(0), color = 'red', linewidth = 10)
    plt.xticks(fontsize=30)
    plt.yticks(fontsize=30)
    plt.xlabel('Iteration', fontsize=30)
    plt.ylabel('Bandgap (eV)', fontsize=30)
    plt.legend(fontsize=30)
    plt.show()

def add_mean_trajectory(feat_type, split, n_sample):

    traj = feat_dict[feat_type][split][n_sample]['traj']
    yerr = traj.std(0)
    plt.fill_between(range(50), traj.mean(0)-yerr, traj.mean(0) + yerr, alpha = 0.5)
    plt.plot(range(50), traj.mean(0), linewidth = 10)

# plot_mean_trajectory('morgan', 'random', '20')

plt.figure(figsize=(10,10))

feat_type, split, n_sample = 'morgan', 'random', '5'
traj = feat_dict[feat_type][split][n_sample]['traj']
print(traj)
yerr = traj.std(0)
plt.fill_between(range(50), traj.mean(0)-yerr, traj.mean(0) + yerr, alpha = 0.5)
plt.plot(range(50), traj.mean(0), linewidth = 10)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.xlabel('Iteration', fontsize=30)
plt.ylabel('Bandgap (eV)', fontsize=30)
plt.legend(fontsize=30)
plt.show()

```
